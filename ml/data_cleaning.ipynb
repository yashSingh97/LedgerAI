### Data Cleaning 

```python
def clean_currency(x):
    if isinstance(x, str):
        # Remove non-numeric characters except decimal and minus
        clean_str = re.sub(r'[^0-9.-]', '', x)
        try:
            return float(clean_str)
        except ValueError:
            return np.nan
    return x
```

```python
# Apply to Rent
if 'Rent' in df.columns:
    df['Rent'] = df['Rent'].apply(clean_currency)
```

```python
# Apply to ALL potential numeric columns that might be object type
numeric_candidates = [
    'Loan_Repayment', 'Insurance', 'Groceries', 'Transport',
    'Eating_Out', 'Entertainment', 'Utilities', 'Healthcare',
    'Education', 'Miscellaneous', 'Disposable_Income', 
    'Desired_Savings', 'Desired_Savings_Percentage'
]
potential_cols = [c for c in df.columns if c.startswith("Potential_Savings")]
numeric_candidates += potential_cols
```

```python
for col in numeric_candidates:
    if col in df.columns:
        # Check if column became object due to noise
        if df[col].dtype == 'object':
            df[col] = df[col].apply(clean_currency)
```

```python
df.info()
```

```python
# Normalize Categorical Columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()
    df[col] = df[col].replace({'nan': np.nan, 'none': np.nan})
```

```python
df['Age'].unique()
```

```python
# clean age
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df = df[df['Age'].between(18, 90)]
```

```python
df['Age'].unique()
```

```python
df['Dependents'].unique()
```

```python
# Clean Dependents
if 'Dependents' in df.columns:
    df['Dependents'] = pd.to_numeric(df['Dependents'], errors='coerce')
    df.loc[df['Dependents'] > 20, 'Dependents'] = np.nan
```

```python
df['Dependents'].unique()
```

```python
# Clean Negatives
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if col not in ['Income', 'Disposable_Income']:
        df[col] = df[col].clip(lower=0)
```

### Feature Engineering

```python
if potential_cols:
    # Fill NaNs with 0 temporarily for summation
    df[potential_cols] = df[potential_cols].fillna(0)
    df["Total_Potential_Savings"] = df[potential_cols].sum(axis=1)
    
    # Remove extreme outliers in Target
    q_high = df["Total_Potential_Savings"].quantile(0.99)
    df = df[df["Total_Potential_Savings"] <= q_high]
    print("Target variable calculated.")
```
